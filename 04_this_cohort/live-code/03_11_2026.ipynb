```python
def eye_exam_covered(age, qualifying_condition, time_since_last_exam):
    '''assume (1) age is passed as an int is for the age of a person
    
    (2) qualifying_condition is passed as a bool (True/False) 
    indicating whether a eligible medical condition is met, 
    
    and (3) time_since_last_exam is passed as an int for the 
    number of months since the last eye exam'''

    if 0 <= age <= 19:# False
        if time_since_last_exam >= 12:
            return "You are eligible for 1 major eye exam and any minor assessments"
        else:
            return "It hasn't been 12 months since your last major eye exam."
    elif 20 <= age <= 64: # True
        if qualifying_condition: # if True:
            if time_since_last_exam >=12: # True
                return "You are eligible for 1 major eye exam and 2 additional follow-up minor assessments"
            else:
                return "It hasn't been 12 months since your last major eye exam"
        else:
            return "You do not have a qualifying eligible medical condition."
    elif age >= 65:
        if time_since_last_exam >= 18:
            return "You are eligible for 1 major eye exam and 2 additional follow-up minor assessments"
        else:
            return "It hasn't been 18 months since your last major eye exam."
    else:
        return "Invalid age input."
```

```python
eye_exam_covered(27, True, 15)
```

```python
eye_exam_covered(19, False, 11)
```

```python
rainbow = {'red', 'orange', 'yellow', 'green', 'blue', 'indigo', 'violet'}
olympic_flag = {'red', 'green', 'yellow', 'blue', 'black'}
```

```python
# difference: find the values in the first set that are not in the second set

rainbow.difference(olympic_flag)
```

```python
olympic_flag.difference(rainbow)
```

```python
# symmetric difference: find values that are in only one of the two sets

rainbow.symmetric_difference(olympic_flag)
```

```python
rainbow
```

```python
olympic_flag
```

```python
# intersection find the values the two sets have in common

rainbow.intersection(olympic_flag)
```

```python
# union combine two sets to get the unique values in both

rainbow.union(olympic_flag)
```
