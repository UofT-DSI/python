```python
def weather_commentary():
    temp = float(input("What is the temperature in degrees C?"))

    if temp <= 0:
        print("It is freezing cold!")
    elif temp <= 15:
        print("It is cold!")
    elif temp <= 30:
        print("It is hot!")
    elif temp <= 45:
        print("It is very hot!")
    else:
        print("It is extremely hot!")
```

```python
weather_commentary()
```
