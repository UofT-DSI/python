```python
# elif short for "else if"

year = 2002

if year >= 2000:
    print('We are in the 21st century.')
elif year >= 1900:
    print('We are in the 20th century.')
elif year >= 1800:
    print('We are in the 19th century.')
else:
    print('We have gone way back in time!')
```

```python
def eye_exam_covered(age, qualifying_condition, time_since_last_exam):
    # assumptions:
    # age int
    # qualifying_condition bool, True if have an eligible medical condition, False otherwise
    # time_since_last_exam int, number of months since last examination

    if 0 <= age <= 19: # False
        if time_since_last_exam >= 12:
            return "You are eligible for 1 major eye exam and any minor assessments needed"
        else:
            return "It hasn't been 12 months since your last major eye exam but you can have a minor assessments"
        
    elif 20 <= age <= 64: # False
        if qualifying_condition: 
            if time_since_last_exam >= 12: 
                return "You are eligible for 1 major eye exam and 2 additional follow-up minor assessments"
            else:
                return "It hasn't been 12 months since your last major eye exam, but you have follow-up minor assessments"
        else:
            return "You do not have an eligible medical condition"
    
    elif age >= 65: # True
        if time_since_last_exam >= 18: # False
            return "You are eligible for 1 major eye exam and 2 follow-up minor assessments"
        else:
            return "It hasn't been 18 months since your last major eye exam, but you have 2 follow-up minor assessments"
    
    else:
        return "Invalid age input."
            
```

```python
eye_exam_covered(-100, True, 5)
```

```python
eye_exam_covered(65, True, 10)
```

```python
eye_exam_covered(50, True, 18)
```

```python
eye_exam_covered(23, False, 10)
```

```python
eye_exam_covered(15, True, 3)
```

```python
eye_exam_covered(19, False, 14)
```
