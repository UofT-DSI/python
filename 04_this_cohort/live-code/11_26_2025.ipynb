```python
def divide(dividend, divisor):
    result = dividend / divisor
    return result
```

```python
divide(0, 2) # 0 is assigned to dividend, 2 assigned to divisor
```

```python
divide(10, 5)
```

```python
divide(2, 0) # 2 dividend, 0 is divisor
```

```python
def calc_total_bill(price, tax_rate=0.13, tip_rate=0.15):
    '''Calculate total bill with tax and tip.'''
    tax = price * tax_rate
    tip = price * tip_rate
    total = price + tax + tip
    return total
```

```python
calc_total_bill(200)
```

```python
calc_total_bill(200, 0.18)
```

```python
calc_total_bill(200, tip_rate=0.18)
```
