```python
def calc_total_bill(price, tax_rate = 0.13, tip_rate = 0.2):
    tax = price * tax_rate
    tip = price * tip_rate
    return price + tax + tip
```

```python
def count_letters(word):
    """
    This function take in a string, and return the number of characters in that string.

    For example, count_letters("hello") would return 5.
    """
    return len(word)
```

```python
count_letters("julia")
```

```python
def rectangle_volume(length, width, height):
    """
    This function will calculate the volume of a rectangle when given a length, width and height.
    """
    return length * width * height
```

```python
rectangle_volume(10,10,5)
```

```python
conversion_rate = 0.73
```

```python
def can_to_us(can):
    """
    Convert canadian dollars to USD.
    Parameters:
    can(float): Amount in CAD
    Returns:
    float: Amount in USD
    """
    return can * conversion_rate
```

```python
can_to_us(10)
```
