## `and`

Evaluates to `True` if both statements are true.

|X|Y|X `and` Y|
|-|-|-|
|True|True|True|
|False|True|False|
|True|False|False|
|False|False|False|

```python
# and: check if the statements on both sides are true

True and False
```

```python
(7 == 7)
```

```python
(9 > 32)
```

```python
(7 == 7) and (9 > 32) # True and False
```

```python
(4 == 2)
```

```python
(5 > 6)
```

```python
(4 == 2) and (5 > 6) # False and False
```

```python
# behaviour of the `and` operator is that it will return the first
# falsy value encountered or the last truthy value  if all operands 
# are true based on the conditions provided
```

```python
is_winter = "True" # string, non-empty and is considered `True`
is_cloudy = True # boolean

is_winter and is_cloudy # "True" and True
```

```python
is_cloudy and is_winter # True and "True"
```

```python
"" and False #empty string and boolean False
```

```python
False and ""
```

```python
0 and 1 # False and True
```

## To avoid confusion and ensure consistent behaviour, it's recommended to use boolean values (`True` or `False`) explicitly when working with logical operations to ensure predictable outcomes based on boolean logic.

## `or`

Evaluates to `True` if just one of the statements is true.

|X|Y|X `or` Y|
|-|-|-|
|True|True|True|
|False|True|True|
|True|False|True|
|False|False|False|

```python
# or: check if at least one of the statements are true

(4 == 4) or False # True or False
```

```python
('Python' == 'python') or True # False or True
```

```python
not (7 % 2 == 1) or False # False or False
```

```python
# or operator return the first truthy value encountered or the last falsy
# value if all operands are `False` based on the conditions provided

True or "True"
```

```python
"True" or True
```

```python
0 or 0
```

```python
0 or ''
```

## Operator precedence

Boolean operators are evaluated after arithmetic and comparison operators.

| Order | Operator | Description |
|---|---|---|
| 1 | `**` | Exponentiation |
| 2 | `-`| Negation |
| 3 | `*`, `/`, `//`, `%` | Multiplication, division, integer division, and modulo |
| 4 | `+`, `-` | Addition and subtraction |
| 5 | `<`, `<=`, `>`, `>=`, `==`, `!=` | Less than, less than or equal to, greater than, greater than or equal to, equal, not equal |
| 6 | `not` | Not |
| 7 | `and` | And |
| 8 | `or` | Or|

```python
year = 2026

if year >= 2000:
    print("We are in the 21st century.")
```
