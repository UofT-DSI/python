Comparison operators

Python has comparison operators as well as arithmetic operators. Unlike arithmetic expressions, which produce integer or float values, comparison expressions produce a Boolean value.

Operator	Description
>	Greater than
>=	Greater than or equal to
<	Less than
<=	Less than or equal to
==	Equal to
!=	Not equal to

```python
-15 >= 14.99
```

```python
30 != 100
```

Operator precedence

Arithmetic and comparison operators are evaluated in the following order. We can enclose operations in parentheses to override the order of precedence -- operations in parentheses are evaluated before the rest of the expression.

Order	Operator	Description
1	**	Exponentiation
2	-	Negation
3	*, /, //, %	Multiplication, division, integer division, and modulo
4	+, -	Addition and subtraction
5	<, <=, >, >=, ==, !=	Less than, less than or equal to, greater than, greater than or equal to, equal, not equal

# Writing our own functions

```python
# start with a definition function "def function_name"
# parameter is a variable used only within a function
# when we pass an argument to a function the argument value is assigned 
# to a corresponding parameter
```

```python
def c_to_f(degrees_c): # write def, then add function name; colon is important to add
    degrees_f = (9 / 5) * degrees_c + 32 # body 
    return degrees_f # return output
# only defined the function here have not called it
```

```python
c_to_f(100) # argument is 100; called the funtion; 100 is being assigned to function
```

```python
c_to_f(-11)
```

```python
freezing_f = c_to_f (0)
```

```python
freezing_f
```

```python
# functions can also have multiple parameters

def divide(dividend, divisor):
    result = dividend / divisor
    return result # if you didn't have a result python would return 'none'
```

```python
divide(0, 2) # 0 is assigned to dividend, and 2 is assigned to divisor
```

```python
divide(2, 0) # gives an error because diving by 0; order matters
```

```python
def calc_sales_tax(price, tax_rate=0.13):
    tax_amt = price * tax_rate
    return tax_amt
```

```python
calc_sales_tax(5) # argument 5 will be assigned to price
```

```python
5 * 0.13
```

```python
calc_sales_tax (5, .08) # .08 is assigned to tax_rate which is the second argument
```

```python
5 * 0.08
```

```python
def calc_total_bill(price, tax_rate=0.13, tip_rate=0.2):
    tax_amt = price * tax_rate
    tip_amt = price * tip_rate
    total_cost = price + tax_amt + tip_rate
    return total_cost
```

```python
calc_total_bill (100)
```

```python
calc_total_bill(100, 0.15)
```

```python
 100 + 100 * 0.15 + 100 * 0.20
```

```python
calc_total_bill(100, tip_rate= 0.15)
```
