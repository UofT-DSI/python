```python
# put together to combine into function
def anagram_checker(word_a, word_b, case_sensitive=False):
    '''
    Check if two strings are anagrams. Insert two strings
    '''
    # convert string to lower case
    if case_sensitive == False:
        lower_a = word_a.lower()
        lower_b = word_b.lower()
    else:
        lower_a = word_a
        lower_b = word_b    
    # break string into word and sort
    string_a = sorted(list(lower_a))
    string_b = sorted(list(lower_b))
    # check similarity
    check = string_a == string_b
    return(check)
```

```python
anagram_checker("Silent", "listen", True) # False
```

```python
anagram_checker("Silent", "Listen", True) # False
```
