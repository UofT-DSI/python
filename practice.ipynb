```python
def calc_total_bill (price, tax_rate = 0.13, tip_rate = 0.2):

    """
    This function calculates the total bill after tip and tax.
    """
    tax = price * tax_rate
    tip = price * tip_rate
    return price + tax + tip
```

```python
calc_total_bill (100)
```

```python
calc_total_bill (100, tip_rate = 0.22)
```
