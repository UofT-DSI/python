```python
dr = pd.read_excel('../data/ttc-subway-delay-codes.xlsx', sheet_name=None)
```

```python
type(dr)
```

```python
delay_reasons = pd.DataFrame()

for sheet_name, sheet_contents in dr.items():
    delay_reasons = pd.concat([delay_reasons, sheet_contents],
                              axis=0,
                              ignore_index=True)

delay_reasons
```

```python
def clean_names(string):
    return string.lower().replace(' ', '_')
```

```python
all_delays.rename(columns={'date':clean_names('date'), })
```

```python
print(list(delay_reasons))
print(list(all_delays))
```

```python
all_delays = all_delays.rename(columns={'DATE':'date'})
```

```python
all_delays['DATE'].rename(clean_names)
```

```python
delay_reasons.head()
```

```python
# left: merge will include all rows from the `all_delays` dataframe and only those rows from 'delay_reasons'
# that have matching values in `all_delays`

# right: resulting dataframe contains all rows from the right dataframe and the matched rows from the 
# left dataframe

# inner: resulting dataframe contains only the rows that have matching keys in both the left and right dataframes

# outer: resulting dataframe includes all rows from both the left and right datagrames, where there are missing
# matches you'll see NaN

delays_w_reasons = pd.merge(all_delays,
                            delay_reasons,
                            how='left', #type of merge to perform
                            left_on='code',
                            right_on='rmenu_code')
delays_w_reasons.head(3)
```

```python
delays_w_reasons = delays_w_reasons.drop(columns='rmenu_code')
delays_w_reasons.head(3)
```
