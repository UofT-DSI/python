```python
def calc_total_bill(price, tax_rate=0.13, tip_rate=0.3):
    """calculate total bill with tax and tip"""
    tax = price * tax_rate
    tip = price * tip_rate
    total = price + tax + tip
    return total
```

```python
calc_total_bill(120)
```

```python
calc_total_bill (120, tip_rate=0.4)
```

```python
calc_total_bill (150, 0.18)
```
