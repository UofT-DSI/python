# Control flow

```python
# conditions
condition = True
if condition:
    # code that runs if condition is true
    print(condition)

# Example of basic if statement
year = 2022
if year > 2000:
    print('We are in the 21st century')
```

```python
# Example of if/else statement
year = 1999
if year >= 2000:
    print('We are in the 21st century')
else:
    print('We are not in the 21st century')
```

```python
# Example of if/elif/else statement
year = 1867
if year > 2000:
    print('We are in the 21st century')
elif year > 1900:
    print('We are in the 20st century')
else:
    print('We are not in the 20th or 21st century')
```

```python
# Exercise
# Write a program where it outputs how cold the weather is based on these
# conditions:
# Less than 0C, it is freezing cold!
# Between 1 and 15C, it is cold!
# Between 16 and 30C, it is hot!
# Between 31 and 45C, it is really hot!
# Over 45C, it is extremely hot!

def output_temp(temp):
    if temp < 0:
        return 'it is freezing cold!'
    elif temp <= 15:
        return 'it is cold!'
    elif temp <= 30:
        return 'it is hot!'
    elif temp <= 45:
        return 'it is really hot!'
    else:
        return 'it is extremely hot!'

input_temp = input("What is the temperature?")

print(output_temp(int(input_temp)))
```

```python
# nested conditions

age = 20
has_id = True

if age >= 18:
    if has_id:
        print('You are able to enter')
    else:
        print('You need an ID')
else:
    print("You are too young")
```
