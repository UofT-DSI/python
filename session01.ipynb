# Exercise - F to C

```python
def f_to_c(degrees_fahrenheit):
    return (degrees_fahrenheit - 32) * (5/9)
```

# Optional params and defaults

```python
def calc_sales_tax(price,tax_rate=0.13):
    return price * tax_rate
```

```python
calc_sales_tax(100)
```

```python
calc_sales_tax(100,.05)
```

# Keyword args

```python
def calc_total_bill(price, tax_rate=0.13, tip_rate=0.2):
    tax = price * tax_rate
    tip = price * tip_rate
    return price + tax + tip
```

```python
calc_total_bill(100)
```

```python
calc_total_bill(100, tip_rate=0.22)
```

```python
# this call assigns 0.22 to tax_rate due to order
calc_total_bill(100,0.22)
```
